```python
df = pd.read_csv('../Kaggle_dataset/Cleaned_2018_Flights.csv')
```

```python
df = df[df['AirlineCompany'].isin(['DL', 'WN', 'UA', 'AA'])]
df = df[['Unnamed: 0','Quarter','Miles','Origin','Dest','NumTicketsOrdered', 'AirlineCompany','PricePerTicket']]
```

```python
app_df = pd.read_excel('Airport Rankings 2018.xlsx',header=2,usecols=[0,1])
```

```python
app_df.iloc[0:5,:]
```

```python
app_df2 = pd.read_excel('Airport Rankings 2018.xlsx',sheet_name=3,header=4,usecols=[0,1,2])
```

```python
app_df2.head()
```

```python
app_df2['Airport'] = app_df2['Airport'].apply(lambda x: x.split(',')[0])
```

```python
app_df2.head()
```

```python
app_df2.info()
```

```python
df_s = app_df.merge(app_df2[['Airport','Code']],how='left',left_on='Airport',right_on='Airport')
```

```python
df_s = df_s.iloc[:-3,:]
df_s
```

```python
map_dict = {Code:Rank for Rank, Code in zip(df_s['2018 Rank'],df_s['Code'])}
```

```python
map_dict2 = {Code:Rank for Rank, Code in zip(app_df2['Unnamed: 0'],app_df2['Code'])}
```

```python
df_sp = df.sample(frac=0.05)
df_sp.info()
```

```python
df_sp = df_sp[['Quarter','Miles','Origin','Dest','NumTicketsOrdered', 'AirlineCompany','PricePerTicket']]
```

```python
df_sp_m = df.replace({"Origin":map_dict2, "Dest":map_dict2})
df_sp_m
```

```python
y = df_sp_m['PricePerTicket']
X = df_sp_m[['Quarter','Miles','Origin','Dest','NumTicketsOrdered']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X.Origin.unique()
```

```python
reg = LinearRegression().fit(X_train, y_train)
reg.score(X_test,y_test)
```
